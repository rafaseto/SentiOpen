```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
labels = ["2.12", "2.13", "2.14", "2.15", "2.16", "2.17", "2.18"]
negative = [262, 172, 85, 155, 218, 351, 39]
neutral = [1767, 1532, 1191, 1411, 1326, 2029, 308]
positive = [413, 323, 212, 276, 298, 670, 62]


x = np.arange(len(labels))  # posies para os grupos
width = 0.25  # largura das barras

fig, ax = plt.subplots(figsize=(10, 6), dpi=150)

# Criao das barras para cada categoria
bar1 = ax.bar(x - width, negative, width, label='Negative', color="salmon", zorder=3)
bar2 = ax.bar(x, neutral, width, label='Neutral', color="silver", zorder=3)
bar3 = ax.bar(x + width, positive, width, label='Positive', color="lightgreen", zorder=3)

# Configuraes do grfico
ax.set_ylabel('Number of Comments', fontsize=14)
ax.set_xlabel('Release', fontsize=14)
ax.set_title('Comment Sentiment Analysis in GitHub Issues Across TensorFlow Releases (via Gemini 2.0 Flash)', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha="right")
ax.legend(fontsize=14)  

# Adicionando grid no eixo y (com zorder menor para ficar atrs das barras)
ax.grid(axis='y', alpha=0.7, zorder=0)

ax.tick_params(axis='x', which='both', bottom=False, top=False)
ax.tick_params(axis='y', which='both', left=False, right=False)

plt.tight_layout()

plt.savefig('gemini_tensorflow_final.png', dpi=300)

plt.show()
```
